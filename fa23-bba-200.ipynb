```c++
from datetime import datetime

def greet_with_time(name):
    current_time = datetime.now().strftime("%H:%M:%S")
    print(f"Hello, {name}! Current time is {current_time}")

greet_with_time("Talha")
```

```c++
def calculate_average(x, y, z):
    return (x + y + z) / 3

print(calculate_average(45, 30, 55))
```

```c++
def calculate_discount(price, discount_percentage=10):
    discount_amount = price * (discount_percentage / 100)
    final_price = price - discount_amount
    return final_price

print(calculate_discount(1000))
print(calculate_discount(1000, 20))
```

```c++
def print_info(name, age, city):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"City: {city}")

print_info(name="Talha", age=20, city="Lahore")
```

```c++
def find_max(*numbers):
    return max(numbers)

print(find_max(4, 8, 15, 2, 20))
```

```c++
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

```c++
def apply_function(func, numbers):
    return [func(num) for num in numbers]

def double(x):
    return x * 2

print(apply_function(double, [1, 2, 3, 4]))
```
